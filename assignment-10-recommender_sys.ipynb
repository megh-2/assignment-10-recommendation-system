```python
user_book = book.pivot_table(index =["user_id"],columns = ["book_title"],values = "book_rating").reset_index(drop=True)
```

```python
user_book
```

```python
# Replacing the index values by unique UserId
user_book.index=book['user_id'].unique()
```

```python
user_book
```

```python
# Impute those NaNs with 0 values
user_book.fillna(0,inplace=True)
user_book
```

```python
# calculating cosine similarity between users

from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances(user_book.values,metric = "cosine")
```

```python
user_sim
```

```python
# store the results in a dataframe

user_sim_df = pd.DataFrame(user_sim)
```

```python
user_sim_df
```

```python
# set the index an d columns names  to user ids
user_sim_df.index = book.user_id.unique()
user_sim_df.columns = book.user_id.unique()
```

```python
def get_recommendation(customer_id):
        
        tem=list(user_sim_df.sort_values([customer_id],ascending=False).head(3).index)

        book_list=list(book[book['user_id'].isin(tem)]['book_title'])

        recommended_books = set(book_list)-set(book[book['user_id']==customer_id]['book_title'])
        
        return recommended_books
```
